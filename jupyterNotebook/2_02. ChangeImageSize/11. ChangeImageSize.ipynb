# [Resize and crop images.]

###  1. Image resizing.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import cv2

#! Read with OpenCV.
img = cv2.imread("test.jpg")

#!
#! Display image.(Original)
#!
print("<< ORIGINAL >>")
#! Delete AXIS numerical scale.
#plt.axis("off") 
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

#!
#! Image resizing.
#! - param2: resize(Specify by tuple.)
img2 = cv2.resize(img, (600, 300))

#! Saving resized images.
cv2.imwrite("out-resize.png", img2)

#! Display image.
print("<< RESIE >>")
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```

###  2. Image cutting.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import cv2

#! Read with OpenCV.
img = cv2.imread("test.jpg")

#!
#! Image cutting.
#! - [y1: y2, x1: x2]
img2 = img[150:450, 150:450]
#! Image resizing.
img2 = cv2.resize(img2, (400, 400))

#! Saving resized images.
cv2.imwrite("out-resize.png", img2)

#! Display image.
print("<< CUT >>")
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.show()
```
