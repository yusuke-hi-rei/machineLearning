### ・Performs negative / positive reversal.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import cv2

img = cv2.imread("test.jpg")
#! Performs negative / positive reversal.
img = 255 - img
#! Show image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### ・Convert to grayscale.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import cv2

img = cv2.imread("test.jpg")
#! Performs negative / positive reversal.
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#! Show image
plt.imshow(img, cmap="gray")
plt.axis("off")
plt.show()
```

### ・Constants often used for color space conversion.

| | 定数 | 効果 |
|---- | :---- | :---- |
| 1 | cv2.COLOR_BGR2GRAY | BGRカラー画像をグレイスケールに変換。|
| 2 | cv2.COLOR_RGB2BGR | RGBカラー画像をBGRカラー画像に変換。|
| 3 | cv2.COLOR_BGR2YCrCb | BGRカラー画像をYCrCbに変換。|
| 4 | cv2.COLOR_BGR2HSV | BGRカラー画像をHSVに変換。|

### ・Reverse rotation of the image.

```python
%matplotlib inline
#!
#! Flip horizontally and flip vertically.
#!
import matplotlib.pyplot as plt
import cv2

#! Load image.
img = cv2.imread("test.jpg")

#! 
#! Specify the graph frame size with figsize (horizontal, vertical)
#! 
plt.figure(figsize=(8, 8)) 

#! Original image left.
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#! Flip image left / right.
plt.subplot(1, 3, 2)
img2 = cv2.flip(img, 1) # left right
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))

#! Flip image upside down.
plt.subplot(1, 3, 3)
img3 = cv2.flip(img, 0) # 0: up down.
plt.imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
%matplotlib inline
#!
#! Image rotation.
#!
import matplotlib.pyplot as plt
import cv2

#! Load image.
img = cv2.imread("test.jpg")

#! 
#! Specify the graph frame size with figsize (horizontal, vertical)
#! 
plt.figure(figsize=(8, 8)) 

#! Get image size.
h, w, color = img.shape 
size = (w, h) # tuple
#print(type(size))

#! Image center position.
center = (w // 2, h // 2) # tuple

#! Get rotation transformation matrix.
angle = 45
scale = 1.0
#! Generate 2D rotation transformation matrix.
matrix = cv2.getRotationMatrix2D(center, angle, scale)
#! アフィン変換
img2 = cv2.warpAffine(img, matrix, size)

#! Original image left.
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#! Rotate image right.
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))

plt.show()
```
